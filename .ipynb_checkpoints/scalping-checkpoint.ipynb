```python
import pandas as pd
import time
```

```python
base_url = "https://djpk.kemenkeu.go.id/portal/data/apbd"
years = list(range(2011, 2026))
periods = [2]

all_data = []
```

```python
for year in years:
    for p in periods:
        target_url = f"{base_url}?periode={p}&tahun={year}&provinsi=--&pemda=--"
        print(f"Fetching: Year {year}, Period {p}...")
        
        try:
            tables = pd.read_html(target_url)
            if tables:
                df = tables[0]
                df['Year'] = year
                all_data.append(df)
            
            time.sleep(3) 
            
        except Exception as e:
            print(f"Could not fetch data for {year} P{p}: {e}")

if all_data:
    final_df = pd.concat(all_data, ignore_index=True)
    final_df.to_csv("indonesia_budget_master.csv", index=False)
    print("Master file created successfully!")
```
