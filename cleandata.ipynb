```python
import pandas as pd
```

```python
# Data Cleaning
data = pd.read_csv("All_data_ocd.csv")

data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')

date_cols = ['date', 'tender_datepublished', 'award_date']

for col in date_cols:
    data[col] = pd.to_datetime(data[col], utc=True, errors='coerce')
    data[col] = data[col].dt.tz_localize(None)

num_cols = ['tender_minvalue', 'award_value']

for col in num_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

data.drop_duplicates(inplace=True)

data.drop(columns=['buyer_id'], inplace=True)

data.dropna(subset=['tender_id', 'award_value', 'award_supplier'], inplace=True)

data['mainprocurementcategory'] = data['mainprocurementcategory'].fillna('unknown')

text_cols = ['tender_title', 'award_title', 'buyer_name', 'award_supplier']

for col in text_cols:
    data[col] = data[col].str.strip()
    data[col] = data[col].str.title()

data.to_csv("All_data_ocd_cleaned.csv", index=False)
```
